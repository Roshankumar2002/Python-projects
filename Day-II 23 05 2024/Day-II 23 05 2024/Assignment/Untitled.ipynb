```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'],bins=10,density=False)
print(counts)
print(bin_edges)
pdf = counts/sum(counts)
print(pdf)
print('the sum of counts is',sum(pdf))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='cdf')
plt.plot(bin_edges[1:],cdf,label='pdf')
plt.xlabel('petal_length_setosa')
plt.ylabel('probabilities')
plt.legend()
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='setosa_PDF')
plt.plot(bin_edges[1:], cdf,label='setosa_CDF')
counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='virginica_PDF')
plt.plot(bin_edges[1:], cdf,label='virginica_CDF')
plt.xlabel('petal_length')
plt.ylabel('probabilities')
plt.legend()

plt.show();
```
